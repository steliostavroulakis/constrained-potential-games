# Hyperparameter Tuning

We used an MLOps platform called Weights and Biases to configure various hyperparameters of our experiment and quantify relationships between some variables and the duality gap:

| Parameter | Range |
| --- | --- |
|`regularizer`|`[0.005, 0.01, 0.02]`|
|`strategy_stepsize`|`[0.001, 0.002, 0.005]`|
|`gas`|`[2, 4, 6, 8, 10]`|
|`hw_cong_mult`|`[0.05, 0.1, 0.2]`|

Each combination of hyperparameter yields a different Nash Gap. Below are the results of the sweep:

# Two interesting cases

We present two interesting runs illustrating how players constrain themselves when the gas amount changes.

## Case 1: Approximately enough gas to select the highway

All agents start selecting the highway to minimize congestion. Nash Gap goes to zero.
